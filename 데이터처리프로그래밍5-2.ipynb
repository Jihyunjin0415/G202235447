```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('C:/Users/jhjjh/Downloads/04_data1.xlsx', index_col='area')
data.head()
```

```python
data2020_good = data['2020_good']
data2019_good = data['2019_good']
```

```python
plt.figure(figsize=(20,4))
plt.plot(data2020_good, color='b', marker='o') 
plt.plot(data2019_good, color='r', marker='o')
plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.title('2020 Fine Dust (pm10) Good Line Graph')
plt.legend()
plt.grid() 
plt.show()
```

```python
plt.figure(figsize=(20,4))

for year in range (2015, 2021):
    year = str(year)+'_good'
    chartdata = data[year]
    plt.plot(chartdata, marker='o', label=year)
plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.title('2015-2020 Fine Dust (pm10) Good Line Graph')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
index = np.arange (7)

plt.figure(figsize=(20,4))

data6 = data.loc['Seoul':'Ulsan', '2015_good':'2020_good']
for year in range (2015, 2021):
    year = str(year)+'_good'
    chartdata = data6[year]
    plt.bar(index, chartdata, width=0.15, label=year)
    index = index + 0.15
    
plt.xlabel('Area')
plt.ylabel ('Micrometer')
plt.xticks(index-0.53, ['Seoul', 'Busan', 'Daegu','Incheon',
'Gwangju', 'Daejeon', 'Ulsan'])
plt.title('2015~2020 Fine Dust(pm10) Good Group Bar Graph')
plt.legend()
plt.show()
```

```python
data = pd.read_csv('C:/Users/jhjjh/Downloads/04_data2.csv')
data.head()
```

```python
PetalLength = data['PetalLength'] 
PetalWidth = data['PetalWidth']
plt.figure(figsize=(10,4))
plt.scatter(PetalLength, PetalWidth)
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Iris Petal Scatter Graph')
plt.grid() 
plt.show()
```

```python
PetalLength = data['PetalLength'] 
PetalWidth = data['PetalWidth']

SepalLength = data['SepalLength'] 
SepalWidth = data['SepalWidth']

plt.figure(figsize=(10,4))
plt.scatter(PetalLength, PetalWidth, color='r', edgecolor='w', label='PetalLength PetalWidth')
plt.scatter (SepalLength, SepalWidth, color='b', edgecolor='w',label='SepalLength * SepalWidth')
plt.xlabel('Length')
plt.ylabel('Width')
plt.title('Iris Petal Group Scatter Graph')
plt.legend()
plt.grid()
plt.show()
```

```python
setosa_SepalWidth = data.loc[data.Species == 'setosa', ['Species','SepalWidth']]
versicolor_SepalWidth = data.loc[data.Species == 'versicolor', ['Species','SepalWidth']]

plt.figure(figsize=(10,6))
plt.hist(setosa_SepalWidth['SepalWidth'], bins = 5, alpha=0.5,label='setosa', color='red')
plt.hist(versicolor_SepalWidth['SepalWidth'], bins = 5, alpha=0.5,label='versicolor',color='blue')

plt.xlabel('Sepal')
plt.ylabel('Width')
plt.legend()
plt.grid()
plt.show()
```
